```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
#My Table:
class VeggieExportDest(Base):
    __tablename__ = 'veggieexportdest'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    country = Column(String)
    share = Column(Float)
    avg = Column(Float)
    yr15 = Column(Float)
    yr16 = Column(Float)
    yr17 = Column(Float)
    yr18 = Column(Float)
    yr19 = Column(Float)
    lat = Column(Float)
    lon = Column(Float)

class VeggieImportSource(Base):
    __tablename__ = 'veggieimportsource'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    country = Column(String)
    share = Column(Float)
    avg = Column(Float)
    yr15 = Column(Float)
    yr16 = Column(Float)
    yr17 = Column(Float)
    yr18 = Column(Float)
    yr19 = Column(Float)
    lat = Column(Float)
    lon = Column(Float)

class VeggieImportValue(Base):
    __tablename__ = 'veggieimportval'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    mktyr = Column(Float)
    avg = Column(Float)
    jan = Column(Float)
    feb = Column(Float)
    mar = Column(Float)
    apr = Column(Float)
    may = Column(Float)
    jun = Column(Float)
    jul = Column(Float)
    aug = Column(Float)
    sep = Column(Float)
    oct = Column(Float)
    nov = Column(Float)
    dec = Column(Float)
    
class VeggieExportValue(Base):
    __tablename__ = 'veggieexportval'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    mktyr = Column(Float)
    avg = Column(Float)
    jan = Column(Float)
    feb = Column(Float)
    mar = Column(Float)
    apr = Column(Float)
    may = Column(Float)
    jun = Column(Float)
    jul = Column(Float)
    aug = Column(Float)
    sep = Column(Float)
    oct = Column(Float)
    nov = Column(Float)
    dec = Column(Float)    

class FruitExportDest(Base):
    __tablename__ = 'fruitexportdest'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    country = Column(String)
    share = Column(Float)
    avg = Column(Float)
    yr15 = Column(Float)
    yr16 = Column(Float)
    yr17 = Column(Float)
    yr18 = Column(Float)
    yr19 = Column(Float)
    lat = Column(Float)
    lon = Column(Float)

class FruitImportSource(Base):
    __tablename__ = 'fruitimportsource'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    country = Column(String)
    share = Column(Float)
    avg = Column(Float)
    yr15 = Column(Float)
    yr16 = Column(Float)
    yr17 = Column(Float)
    yr18 = Column(Float)
    yr19 = Column(Float)
    lat = Column(Float)
    lon = Column(Float)

class FruitImportValue(Base):
    __tablename__ = 'fruitimportval'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    mktyr = Column(Float)
    avg = Column(Float)
    jan = Column(Float)
    feb = Column(Float)
    mar = Column(Float)
    apr = Column(Float)
    may = Column(Float)
    jun = Column(Float)
    jul = Column(Float)
    aug = Column(Float)
    sep = Column(Float)
    oct = Column(Float)
    nov = Column(Float)
    dec = Column(Float)
    
class FruitExportValue(Base):
    __tablename__ = 'fruitexportval'
    id = Column(Integer, primary_key=True)
    product = Column(String)
    flow = Column(String)
    mktyr = Column(Float)
    avg = Column(Float)
    jan = Column(Float)
    feb = Column(Float)
    mar = Column(Float)
    apr = Column(Float)
    may = Column(Float)
    jun = Column(Float)
    jul = Column(Float)
    aug = Column(Float)
    sep = Column(Float)
    oct = Column(Float)
    nov = Column(Float)
    dec = Column(Float)    
```

```python
engine = create_engine('sqlite:///Veggie_Fruit_DB.sqlite')
```

```python
Base.metadata.create_all(engine)
```

```python
import requests
import json
import pandas as pd
import numpy as np
from pprint import pprint
```

```python
csv_file = "Raw data/Veggie Export Destination.csv"
veggie_export_dest = pd.read_csv(csv_file)
```

```python
veggie_export_dest=veggie_export_dest.reset_index(drop=True)
veggie_export_dest.head()
```

```python
veggie_export_dest["lat"] = ""
veggie_export_dest["lon"] = ""
veggie_export_dest.head()
```

```python
gkey = "AIzaSyD7O3afaWLj_VvnQFXQ3ESdH84-TNJ98DI"
for index,row in veggie_export_dest.iterrows():
    try:
        country = row["country"]
        target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
        'address={0}&key={1}').format(country, gkey)
        geo_data = requests.get(target_url).json()
        lat = geo_data["results"][0]["geometry"]["location"]["lat"]
        lon = geo_data["results"][0]["geometry"]["location"]["lng"]
        veggie_export_dest.loc[index,"lat"] = lat
        veggie_export_dest.loc[index,"lon"] = lon
        
    except KeyError:
        veggie_export_dest.loc[index,"lat"] = np.NaN
        veggie_export_dest.loc[index,"lon"] = np.NaN
        
veggie_export_dest.head()
```
